# SIMPLE LINEAR REGRESSION 
## Basic level problem using Scikit Learn Pipelines 
1) Use input data and output data
2) Split the data into train and  test
3) choose the regression model
4) train the model and predict for output
5) Compare the data with already existing values to find error
### Use of Matplotlibrary :
1) Can be used to plot existing datasets
2) To plot the best fit line of linear regression in the original dataset

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([
    [600], [750], [900], [1000], [1100],
    [1250], [1400], [1500], [1650], [1800],
    [2000], [2200], [2500], [2800], [3000]
])

y = np.array([
    32, 38, 48, 55, 63,
    75, 89, 100, 118, 135,
    160, 185, 225, 270, 320
])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

## To find errors in the model 


```python
mse=mean_squared_error(y_test,y_pred)
print(f"MSE = ",mse)
print()
r2=r2_score(y_test,y_pred)
print("r2 = ",r2)
```

```python


# Predict for full X for smooth line
y_pred_line = model.predict(x)

plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred_line, color='red', label='Regression Line')

plt.xlabel("Area (Sqft)")
plt.ylabel("Price (Lakhs)")
plt.title("Sqft vs Price (Linear Regression)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train) # use to train the regression model
```

```python
y_pred=model.predict(x_test) # usewd for predicting the regression model 
```
